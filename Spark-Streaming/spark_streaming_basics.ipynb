# Some commands

Installing Spark:

`sdk install spark`

Starting the Spark cluster with master and 1 worker:

`sbin/start-master.sh --webui-port 8480`

`sbin/start-worker.sh spark://<ip-addr or hostname>:7077`

Filling a kafka topic with stream from a public data source:

`nc 153.44.253.27 5631 | gpsdecode | kafkacat -b 172.17.0.1:49153 -t ais-test`

```python
df = spark.readStream \
    .format("org.apache.spark.sql.kafka010.KafkaSourceProvider") \
    .option("kafka.bootstrap.servers", "172.17.0.1:49153") \
    .option("subscribe", "ais-test") \
    .option("startingOffset", "earliest") \
    .load()
```

```python
df.writeStream \
  .format("console") \
  .outputMode("append") \
  .start() \
  .awaitTermination()
```
